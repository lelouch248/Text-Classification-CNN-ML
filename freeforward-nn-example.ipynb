```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import RandomOverSampler
```

```python
df = pd.read_csv("/content/drive/MyDrive/Data Science/Tf_textclassification_data/Copy of diabetes.csv")
df
```

```python
X = df[df.columns[:-1]].values
y = df[df.columns[-1]].values
```

```python
scaler = StandardScaler()
X  = scaler.fit_transform(X)
data = np.hstack((X,np.reshape(y,(-1,1))))
transformed_df = pd.DataFrame(data,columns=df.columns)
```

```python
for i in range(len(transformed_df.columns[:-1])):
  label = transformed_df.columns[i]
  plt.hist(transformed_df[transformed_df['Outcome']==1][label], color = 'blue',label = "diabetes",alpha = 0.7,density = True,bins = 15)
  plt.hist(transformed_df[transformed_df['Outcome'] == 0][label],color = 'red',label = "no diabetes",alpha = 0.7,density = True,bins = 15)
  plt.title(label)
  plt.ylabel("probablity")
  plt.xlabel(label)
  plt.legend()
  plt.show()

```

```python
over = RandomOverSampler()
X,y = over.fit_resample(X,y)
data = np.hstack((X,np.reshape(y,(-1,1))))
transformed_df = pd.DataFrame(data,columns=df.columns)
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtemp,ytrain,ytemp = train_test_split(X,y,test_size = 0.4,random_state = 0)
xvalid,xtest,yvalid,ytest = train_test_split(xtemp,ytemp,test_size = 0.5,random_state = 0)
```

```python
model = tf.keras.Sequential([
                             tf.keras.layers.Dense(16, activation = 'relu'), #if x is less than zero it is zero and if x is greater than zero then it is just x
                             tf.keras.layers.Dense(16, activation = 'relu'),
                             tf.keras.layers.Dense(1, activation = 'sigmoid')
                        
])
```

```python
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
              loss = tf.keras.losses.BinaryCrossentropy(),
              metrics = ['accuracy'])
```

```python
model.evaluate(xtrain,ytrain)
```

```python
model.fit(xtrain,ytrain,batch_size = 16,epochs=30,validation_data = (xvalid,yvalid))
```

```python
model.evaluate(xtest,ytest)
```
